```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector, plot_state_qsphere
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(1, 1) # Quantum Circuit with 1 qubit and 1 classical bit.
circuit.x(0)  # Apply the Pauli-X gate.
circuit.h(0) # Apply the Hadamard gate.
print(circuit)
circuit.draw(output = 'mpl') # Draw the circuit diagram.
```

```python
simulator = Aer.get_backend('statevector_simulator') # Simulate the QC execution.
result = execute(circuit, backend=simulator).result()
state_vector = result.get_statevector()
array_to_latex(state_vector, prefix="\\text{State Vector = }\n") # Visual representation of the vector.
```

```python
plot_bloch_multivector(state_vector) # Vector visualization on the bloch sphere.
```

```python
plot_state_qsphere(state_vector) # Vector visualization on the bloch sphere.
```

```python
circuit.measure(0, 0)
circuit.draw(output = 'mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator') # Simulate the QC execution.
result = execute(circuit, backend=simulator, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```
