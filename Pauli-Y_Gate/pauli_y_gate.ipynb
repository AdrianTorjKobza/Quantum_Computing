```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(1) # Quantum Circuit with 1 qubit.
circuit.y(0) # Apply Pauli-Y gate to the first and only qubit.
circuit.y(0)  # Apply Pauli-Y gate.
circuit.y(0)  # Apply Pauli-Y gate.
print(circuit)
circuit.draw(output = 'mpl') # Draw the circuit diagram.
```

```python
simulator = Aer.get_backend('statevector_simulator') # Simulate the QC execution.
result = execute(circuit, backend=simulator).result()
state_vector = result.get_statevector()
array_to_latex(state_vector, prefix="\\text{State Vector = }\n") # Visual representation of the vector.
```

```python
plot_bloch_multivector(state_vector) # Vector visualization on the bloch sphere.
```

```python
circuit = QuantumCircuit(2)  # Quantum Circuit with two qubit.
circuit.ry(pi/4, [0, 1])  # Apply Pauli-Y gate for the qubit in superposition (different from 0, 1).
circuit.y(1)  # Apply Pauli-Y gate to the second qubit.
print(circuit)
circuit.draw(output='mpl')  # Draw the circuit diagram.
```

```python
result = execute(circuit, backend=simulator).result()
state_vector = result.get_statevector()
array_to_latex(state_vector, prefix="\\text{State Vector = }\n") # Visual representation of the vector.
```

```python
plot_bloch_multivector(state_vector) # Vector visualization on the bloch sphere.
```
