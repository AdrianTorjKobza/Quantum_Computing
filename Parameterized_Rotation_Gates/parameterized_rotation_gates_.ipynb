```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector, plot_state_qsphere
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3) # Quantum Circuit with 3 qubits.
circuit.rx(pi/3, 0) # Apply X-Rotation gate for the first qubit.
circuit.ry(-pi/3, 1)  # Apply Y-Rotation gate for the second qubit.
circuit.h(2) # Apply Hadamard gate for the third qubit.
circuit.rz(2, 2)  # Apply Z-Rotation gate for the third qubit (2 radians = ~115 degrees).
print(circuit)
circuit.draw(output = 'mpl') # Draw the circuit diagram.
```

```python
simulator = Aer.get_backend('statevector_simulator') # Simulate the QC execution.
result = execute(circuit, backend=simulator).result()
state_vector = result.get_statevector()
array_to_latex(state_vector, prefix="\\text{State Vector = }\n") # Visual representation of the vector.
```

```python
plot_bloch_multivector(state_vector) # Vector visualization on the bloch sphere.
```
